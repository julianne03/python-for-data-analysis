```python
import pandas as pd
import numpy as np
```

### Reindexing
- create a new object with the values rearranged to align with the new index

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=["d", "b", "a", "c"])
obj
```

```python
# reindex - rearranges the data according to the new index
obj2 = obj.reindex(["a", "b", "c", "d", "e"])
obj2
```

```python
frame = pd.DataFrame(np.arange(9).reshape((3,3)),
                    index=["a", "c", "d"],
                    columns=["Ohio", "Texas", "California"])
frame
```

```python
frame2 = frame.reindex(index=["a", "b", "c", "d"])
frame2
```

```python
# columns can be reindexed with the columns keyword
states = ["Texas", "Utah", "California"]
frame.reindex(columns=states)
```

```python
# we can also reindex by using loc operator (preferable way)
# works only if all of the new index lables already exist in the DataFrame
frame.loc[["a", "d", "c"], ["California", "Texas"]]
```

### Dropping Entries from an Axis
- `drop` method will return a new object with the indicated value or values deleted from an axis

```python
obj = pd.Series(np.arange(5.), index=["a", "b", "c", "d", "e"])
obj
```

```python
# dropping the row with index of "c"
new_obj = obj.drop("c")
new_obj
```

```python
obj.drop(["d", "c"])
```

```python
data = pd.DataFrame(np.arange(16).reshape((4, 4)),
            index=["Ohio", "Colorado", "Utah", "New York"],
            columns=["one", "two", "three", "four"])
data
```

```python
# drop values from the row labels (axis 0)
data.drop(index=["Colorado", "Ohio"])
```

```python
# drop labels from the columns, instead use the column keyword
data.drop(columns=["two"])
```

### Indexing, Selection, and Filtering

### Selection on DataFrame with loc and iloc
- `loc`: label-based indexing
- `iloc`: integer-based indexing

```python
# select a single row by label
data.loc["Colorado"]
```

```python
# select multiple rows, pass a sequence of labels
data.loc[["Colorado", "New York"]]
```

```python
# combine both row and column selection
# by separating the selections with a comma
data.loc["Colorado", ["two", "three"]]
```

```python
# select a single row by integer
data.iloc[2]
```

```python
data.iloc[[2, 1]]
```

```python
data.iloc[2, [3, 0, 1]]
```

```python
data.iloc[[1, 2], [3, 0, 1]]
```

### Function Application and Mapping

```python
frame = pd.DataFrame(np.random.standard_normal((4, 3)),
            columns=list("bde"),
            index=["Utah", "Ohio", "Texas", "Oregon"])

frame
```

```python
np.abs(frame)
```

```python
# computes the difference between the maximum and minimum of a Series
def f1(x):
    return x.max() - x.min()

# invoked once on each column in frame
frame.apply(f1)
```

```python
# invoked once per row instead - apply across the columns
frame.apply(f1, axis="columns")
```

```python
def my_format(x):
    return f"{x:.2f}"

# apply an element-wise function
frame["e"].map(my_format)
```

### Sorting and Ranking

```python
obj = pd.Series(np.arange(4), index=["d", "a", "b", "c"])
obj
```

```python
# sort object by index
obj.sort_index()
```

```python
# with DataFrame, we can sort by index on either axis
frame = pd.DataFrame(np.arange(8).reshape((2, 4)),
                    index=["three", "one"],
                    columns=["d", "a", "b", "c"])
frame
```

```python
frame.sort_index()
```

```python
frame.sort_index(axis="columns")
```

```python
# descending order
frame.sort_index(axis="columns", ascending=False)
```

```python
# sort a Series by its values
obj = pd.Series([4, 7, -3, 2])
obj.sort_values()
```

```python
# missing values will be sorted to the end of the Series
obj = pd.Series([4, np.nan, 7, np.nan, -3, 2])
obj.sort_values()
```

```python
# missing values can be sorted to the start
obj.sort_values(na_position="first")
```

```python
frame = pd.DataFrame({"b": [4, 7, -3, 2], "a": [0, 1, 0, 1]})
frame
```

```python
frame.sort_values("b")
```

```python
# to sort by multiple columns, pass a list of names
frame.sort_values(["a", "b"])
```

```python
# rank method breaks ties by assigning each group the mean rank
obj = pd.Series([7, -5, 7, 4, 2, 0, 4])
obj.rank()
```

```python
# ranks can also be assigned according to the order in which they're observed
# without breaking the tie
obj.rank(method="first")
```

```python
# rank - descending order
obj.rank(ascending=False)
```

```python
frame = pd.DataFrame({"b": [4.3, 7, -3, 2], "a": [0, 1, 0, 1], "c": [-2, 5, 8, -2.5]})
frame
```

```python
# ranks over the columns
frame.rank(axis="columns")
```
