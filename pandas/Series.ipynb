```python
import pandas as pd
```

## pandas Data Structures

### Series
- **one-dimensional array-like object** containing a sequence of values of the same type
- it also include an associated array of data labels, called its **index**

```python
# shows the index on the left and the values on the right
# default value of index: 0 to N-1 (N is the length of the data)
obj = pd.Series([4, 7, -5, 3])
obj
```

```python
# The data values should be the same type
obj.dtype
```

```python
obj.array
```

```python
obj.index
```

```python
# create a Series with an indentified index
obj2 = pd.Series([4, 7, -5, 3], index=["d","b","a","c"])
obj2
```

```python
obj2["a"]
```

```python
obj2[["c", "a", "d"]]
```

```python
obj2[obj2 > 0]
```

```python
obj2 * 2
```

```python
# You can also think of Series as oredered dictionary,
# as it is a mapping of index values to data values
"b" in obj2
```

```python
# create a Series from it by passing the dictionary
sdata = {"Ohio": 35000, "Texas": 71000, "Oregon": 16000, "Utah": 5000}
obj3 = pd.Series(sdata)
obj3
```

```python
# A Series can be converted back to a dictionary
obj3.to_dict()
```

```python
# override the keys of dictionary by passing an index with the dictionary keys
states = ["California", "Ohio", "Oregon", "Texas"]
obj4 = pd.Series(sdata, index=states)
# no value for "California" -> NaN, "Utah" was not included in states -> excluded
obj4
```

```python
# to detect missing (null, NA) data
pd.isna(obj4)
```

```python
pd.notna(obj4)
```

```python
# instance methods of isna
obj4.isna()
```

```python
# we can perform arithmetic operations with index label
obj3 + obj4
```

```python
# both the Series object itself and its index have a name attibute
obj4.name = "population"
obj4.index.name = "state"
obj4
```
