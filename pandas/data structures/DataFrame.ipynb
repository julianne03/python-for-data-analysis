```python
import pandas as pd
```

## pandas Data Structure

### DataFrame
- a **rectangular table** of data and contains an ordered, named **collection of columns**
- each of columns can be a different value type
- has both a row and column index
- it can be thought of as a dictionary of Series all sharing the same index

```python
data = {"state": ["Ohio", "Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
        "year": [2000, 2001, 2002, 2001, 2002, 2003],
        "pop": [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}

frame = pd.DataFrame(data)
# this will have its index assigned automatically
# the columns are placed according to the order of the keys in data
frame
```

```python
# selects only the first five rows
frame.head()
```

```python
# returns the last five rows
frame.tail()
```

```python
# if you specify the order of columns, it will be arranged in that order
pd.DataFrame(data, columns=["year", "state", "pop"])
```

```python
# if you pass a column that isn't contained in the dictionary, 
# it will appear with missing values
frame2 = pd.DataFrame(data, columns=["year", "state", "pop", "debt"])
frame2
```

### Accessing a column

```python
# a column can be accessed by dictionary-like notation
frame2["state"]
```

```python
# or by using the dot attribute notation
frame2.year
```

### Accessing a row

```python
# rows can be retrieved by position or name with the special iloc and loc attributes
frame2.loc[1] # selects a row by index label
```

```python
# integer location
frame2.iloc[2] # selects a row by position (numbers)
```

```python
# columns can be modified by assignment
frame2["debt"] = 16.5
frame2
```

```python
# when you assign lists or arrays to a column, the value's length must match
# inserting missing values in any index values not present
val = pd.Series([-1.2, -1.5, -1.7], index=[2, 4, 5])
frame2["debt"] = val
frame2
```

```python
# assigning a column that doesn't exist will create a new column
frame2["eastern"] = frame2["state"] == "Ohio"
frame2
```

```python
# to remove a column
del frame2["eastern"]
frame2.columns
```

```python
populations = {"Ohio": {2000: 1.5, 2001: 1.7, 2002: 3.6},
               "Nevada": {2001: 2.4, 2002: 2.9}}
```

```python
# if the nested dictionary is passed,
# outer dictionary keys -> columns, inner keys -> row indices
frame3 = pd.DataFrame(populations)
frame3
```

```python
# transpose the DataFrame
frame3.T
```

```python
pd.DataFrame(populations, index=[2001, 2002, 2003])
```
